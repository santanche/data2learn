```sql
DROP TABLE IF EXISTS synthetic_ehr;

CREATE TABLE synthetic_ehr (
    patient_id  VARCHAR(6)  NOT NULL,
    ehr_id      VARCHAR(8)  NOT NULL,
    disease     VARCHAR(64) NOT NULL,
    symptom     VARCHAR(64) NOT NULL
) AS SELECT patient_id, ehr_id, disease, symptom
FROM CSVREAD('synthetic_ehr.csv');
```

```sql
SELECT * FROM synthetic_ehr ORDER BY ehr_id;
```

```sql
DROP VIEW IF EXISTS ehr;

CREATE VIEW ehr AS
SELECT DISTINCT ehr_id, patient_id, disease
FROM synthetic_ehr;
```

```sql
CALL CSVWRITE('../02-digital-patient/ehr.csv', 'SELECT * FROM ehr');
```

```sql
DROP VIEW IF EXISTS ehr_record;

CREATE VIEW ehr_record AS
SELECT ehr_id, symptom
FROM synthetic_ehr;
```

```sql
CALL CSVWRITE('../02-digital-patient/ehr_record.csv', 'SELECT * FROM ehr_record');
```

```sql
DROP VIEW IF EXISTS patient;

CREATE VIEW patient AS
SELECT DISTINCT patient_id
FROM synthetic_ehr;
```

```sql
SELECT * FROM patient;
```

```sql
CALL CSVWRITE('../02-digital-patient/patient.csv', 'SELECT * FROM patient');
```
