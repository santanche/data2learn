```sql
DROP TABLE IF EXISTS ehr_records;

CREATE TABLE synthetic_ehr (
    patient_id  VARCHAR(6)  NOT NULL,
    ehr_id      VARCHAR(8)  NOT NULL,
    disease     VARCHAR(64) NOT NULL,
    symptom     VARCHAR(64) NOT NULL
) AS SELECT patient_id, ehr_id, disease, symptom
FROM CSVREAD('synthetic_ehr.csv');
```

```sql
SELECT * FROM synthetic_ehr ORDER BY ehr_id;
```

```sql
SELECT DISTINCT ehr_id, patient_id, disease, count(*)
FROM synthetic_ehr
GROUP BY ehr_id, patient_id, disease
ORDER BY ehr_id;
```

```sql
CREATE VIEW ehr_record AS
SELECT ehr_id, symptom
FROM synthetic_ehr;
```
