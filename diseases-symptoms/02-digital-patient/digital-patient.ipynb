```sql
DROP TABLE IF EXISTS ehr_records;

CREATE TABLE ehr_records (
    patient_id  VARCHAR(6)  NOT NULL,
    ehr_id      VARCHAR(8)  NOT NULL,
    disease     VARCHAR(64) NOT NULL,
    symptom     VARCHAR(64) NOT NULL
) AS SELECT patient_id, ehr_id, disease, symptom
FROM CSVREAD('synthetic_ehrs.csv');
```

```sql
DROP VIEW disease_occurrences;

CREATE VIEW disease_occurrences AS
SELECT disease, count(*) occurrences
FROM ehr_records
GROUP BY disease;
```

```sql
SELECT * FROM disease_occurrences;
```

```sql
DROP VIEW symptom_occurrences;

CREATE VIEW symptom_occurrences AS
SELECT disease, symptom, count(*) occurrences
FROM ehr_records
GROUP BY disease, symptom
ORDER BY disease, symptom;
```

```sql
SELECT * FROM symptom_occurrences;
```

```sql
SELECT d.disease, s.symptom, ROUND(CAST(s.occurrences AS DOUBLE)/CAST(d.occurrences AS DOUBLE),3) as importance
FROM disease_occurrences d, symptom_occurrences s
WHERE d.disease = s.disease;
```
