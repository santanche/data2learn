# Parte 1 - Prontuários, Sintomas e Doenças

## Pacientes

Tabela de pacientes, com código e nome do paciente.

```sql
DROP TABLE IF EXISTS patient;

CREATE TABLE patient (
    patient_id  VARCHAR(6)  NOT NULL,
    name        VARCHAR(80) NOT NULL,
    PRIMARY KEY 
) AS SELECT patient_id, name
FROM CSVREAD('patient.csv');
```

```sql
SELECT * FROM patient;
```

## Prontuário (EHR - Electronic Health Record)

Tabela com prontuários eletrônicos simplificados de pacientes, com código do prontuário, código do paciente e doença que foi causa do atendimento.

```sql
DROP TABLE IF EXISTS ehr;

CREATE TABLE ehr (
    ehr_id      VARCHAR(8)  NOT NULL,
    patient_id  VARCHAR(6)  NOT NULL,
    disease     VARCHAR(64) NOT NULL,
) AS SELECT ehr_id, patient_id, disease
FROM CSVREAD('ehr.csv');
```

```sql
SELECT * FROM ehr;
```

## Registros de Prontuário

Registros de sintomas observados em cada prontuário. Como um prontuário pode registrar vários sintomas, esta tabela pode ter vários registros para o mesmo prontuário. Campos: código do prontuário e sintoma.

```sql
DROP TABLE IF EXISTS ehr_record;

CREATE TABLE ehr_record (
    ehr_id      VARCHAR(8)  NOT NULL,
    symptom     VARCHAR(64) NOT NULL
) AS SELECT ehr_id, symptom
FROM CSVREAD('ehr_record.csv');
```

```sql
SELECT * FROM ehr_record;
```

Cada par doença e sintoma é uma tupla da tabela, e a coluna `occurrences` registra quantas vezes aquele sintoma apareceu associado àquela doença, considerando todos os prontuários.

## Tarefa 3

Cada par doença e sintoma é uma tupla da tabela, e a coluna `importance` registra a importância do sintoma para a doença. Ela é calculada dividindo-se o número de vezes que aquele sintoma apareceu associado àquela doença pelo número de vezes que a doença apareceu ao todo.
